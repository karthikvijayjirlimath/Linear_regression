# Linear Regression using Least square error

```python
from matplotlib import pyplot as plt
import numpy as np
x = np.array([1,2,4,3,5])
y = np.array([1,3,3,2,5])

plt.scatter(x,y)
plt.show()
```

y = b0+b1*x

```python
# Estiimating the value of B1

b1 = np.sum((x-np.mean(x))*(y-np.mean(y)))/np.sum((x-np.mean(x))*(x-np.mean(x)))
b0 = np.mean(y) - b1*np.mean(x)
print(b1,b0)
```

```python
#prediction made
p = b1*x+b0
plt.plot(x,p,marker='o',color='g')
plt.scatter(x,y)
plt.legend(['prediction','data'])
plt.show()
```

```python
#Estimating root mean Square Error

rmse = np.sqrt(np.sum((p-y)**2)/len(x))
print(rmse)
```

# Linear Regression using stochastic gradient Descent

```python
# initially consider the follow

# y = b0 + b1 * x
y = y.astype('float16')
x = x.astype('float16')
print(y,x)

p = np.zeros(5)
# error = p[0] - y[0]
# error

# #updating the weights
# alpha = 0.01 #learning rate
# b0 = b0 - alpha * error
# b0
# b1 = b1 - alpha * error
# b1
b0, b1 = 0.0,0.0
def iterating(p,y,alpha=0.01):
    global b0,b1
    for P,Y,X in zip(p,y,x):
        P = b0 + b1 * X
        error = P - Y
        b0 = b0 - alpha * error
        b1 = b1 - alpha * error * X
    return b0,b1

epochs = 30
err = []
while (epochs:= epochs-1) > 0:
    B0,B1 = iterating(p,y,0.0121416181)
    p = B0 + B1 * x
    err.extend(p-y)

plt.plot(range(1,len(err[::2]+1)),err[::2])
plt.grid()
```

```python
plt.scatter(x,y,color='g',marker='o')
plt.plot(x,p,marker='*')
plt.grid()
```

```python
#Calculating error
rmse = np.sqrt(np.sum((p-y)**2)/len(x))
rmse
accuracy = 1.0 - rmse
accuracy
```
